```python
import numpy as np
import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from glob import glob
import sys
import os
import seaborn as sns

sys.path.insert(1, '../src')
from model import *
from wave_loader import *

sns.set()
root = '..'
```

```python
t_loss_cnn_1000 = np.load(os.path.join(root, 'results', 'CNN_Train1000', 'train_losses.npy'))
v_loss_cnn_1000 = np.load(os.path.join(root, 'results', 'CNN_Train1000', 'val_losses.npy'))

t_loss_dense_1000 = np.load(os.path.join(root, 'results', 'Dense_Train1000', 'train_losses.npy'))
v_loss_dense_1000 = np.load(os.path.join(root, 'results', 'Dense_Train1000', 'val_losses.npy'))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
# semilog plot
ax[0].semilogy(t_loss_cnn_1000, label='Train')
ax[0].semilogy(np.arange(1, 5000, 10), v_loss_cnn_1000, label='Validation')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Loss')
ax[0].set_title('CNN')
ax[0].legend()
ax[1].semilogy(t_loss_dense_1000, label='Train')
ax[1].semilogy(np.arange(1, 5000, 10), v_loss_dense_1000, label='Validation')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Loss')
ax[1].set_title('Dense')
ax[1].legend()
plt.show()

```

```python
# plot validation losses
plt.semilogy(np.arange(1, 1000, 10), v_loss_cnn_1000, label='CNN', color='r')
plt.semilogy(np.arange(1, 1000, 10), v_loss_dense_1000, label='Dense', color='b')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('Test Loss')
plt.legend()
```
